```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Exemplo de lista de tags de entrada
tags_entrada = [
    'machine learning',
    'data science',
    'python',
    'neural networks',
    'data analysis'
]

# Exemplo de tags previamente usadas para treinar o modelo
tags_treinamento = [
    'machine learning',
    'data science',
    'python',
    'neural networks',
    'data analysis',
    'deep learning',
    'big data',
    'statistics'
]

# Codificando as tags usando TF-IDF
vectorizer = TfidfVectorizer()
X_treinamento = vectorizer.fit_transform(tags_treinamento)
X_entrada = vectorizer.transform(tags_entrada)

# Treinando o modelo KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_treinamento)

# Calculando a distância das amostras aos centróides
distancias = kmeans.transform(X_entrada)

# Normalizando as distâncias para obter as pontuações de pertencimento entre 0 e 1
pontuacoes_pertencimento = 1 / (1 + distancias)

# Imprimindo as pontuações de pertencimento para cada amostra
for i, tag in enumerate(tags_entrada):
    print(f'Tag: {tag} - Pontuações de Pertencimento aos Clusters: {pontuacoes_pertencimento[i]}')
```

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
 
import matplotlib.pyplot as plt
#Load Data
data = load_digits().data
pca = PCA(2)
 
#Transform the data
df = pca.fit_transform(data)
 
#Import KMeans module
from sklearn.cluster import KMeans
 
#Initialize the class object
kmeans = KMeans(n_clusters= 10)
 
#predict the labels of clusters.
label = kmeans.fit_predict(df)
 
#Getting unique labels
u_labels = np.unique(label)
 
#plotting the results:
for i in u_labels:
    plt.scatter(df[label == i , 0] , df[label == i , 1] , label = i)
plt.legend()
plt.show()
```
